Basic method

```python
import pandas as pd
data = {'Name': ['Pooja', 'Rahul', 'Rohit'], 'Age': [23,18,20], 'city': ['Mummbai', 'Pune', 'Nagpur']}
data1 = ['Apple', 'Orange', 'Grapes']
# for one dimensional we use 'Series'
dp = pd.Series(data1)
print(dp)
# for multi dimensional we use 'DataFrame'
df = pd.DataFrame(data)
print(df)
# Accessing a specific column
print('Print single column')
print(df['Name'])
# Accessing a specific row
print("Print single row")
print(df.iloc[1])
# Accessing a specific element
print("Print single element")
print(df.loc[2, 'Age'])
print(df.at[2, 'Name'])
# Dropping rows with missin values
print("Dropping rows with missing values")
df.dropna(inplace=True)
print(df)
# Dropping columns with missing values
print("Dropping columns with missing values")
df.dropna(axis=1, inplace=True)
print(df)
# Filling missing values with a specific value
print("Filling missing values with a specific value")
df.fillna(value=0, inplace=True)
print(df)
# Filling missing values with the mean of the column
print("Filling missing values with the mean of the column")
df.fillna(value=df['Age'].mean(), inplace=True)
print(df)
# Filling missing values with the median of the column
print("Filling missing values with the median of the column")
df.fillna(value=df['Age'].median(), inplace=True)
print(df)
# Filling missing values with the mode of the column
print("Filling missing values with the mode of the column")
df.fillna(value=df['Age'].mode()[0], inplace=True)
print(df)
# Grouping data
avg_age = df.groupby('Name')['Age'].mean()
print(avg_age)
```

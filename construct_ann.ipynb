```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
dt = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Admission%20Chance.csv')
dt.head()
```

```python
dt.info()
```

```python
dt.describe()
```

```python
dt.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(dt[['GRE Score', 'TOEFL Score',
       'University Rating', ' SOP', 'LOR ', 'CGPA', 'Research']],
       dt['Chance of Admit '], test_size=0.3, random_state=2529)
```

```python
model = Sequential()
```

```python
# input layer
model.add(Dense(7, activation='relu'))
```

```python
# Hidden layer
model.add(Dense(14, activation='relu'))
```

```python
# output layer
model.add(Dense(1, activation='linear'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model
loss, mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
print(f"Test MAE: {mae}")
```
